```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def mod_inverse(a, mod):
    return pow(a, mod - 2, mod)  # Using Fermat's Little Theorem

def fft(a, invert, mod=7340033, root=5, root_1=4404020, root_pw=1 << 20):
    n = len(a)
    
    for i in range(1, n):
        j = 0
        bit = n >> 1
        while j & bit:
            j ^= bit
            bit >>= 1
        j ^= bit
        if i < j:
            a[i], a[j] = a[j], a[i]
    
    length = 2
    while length <= n:
        wlen = root_1 if invert else root
        i = length
        while i < root_pw:
            wlen = (wlen * wlen) % mod
            i <<= 1
        
        for i in range(0, n, length):
            w = 1
            for j in range(length // 2):
                u = a[i + j]
                v = (a[i + j + length // 2] * w) % mod
                a[i + j] = (u + v) % mod
                a[i + j + length // 2] = (u - v + mod) % mod
                w = (w * wlen) % mod
        
        length <<= 1
    
    if invert:
        n_inv = mod_inverse(n, mod)
        for i in range(n):
            a[i] = (a[i] * n_inv) % mod

# Load image
img = cv2.imread('p5.jpg', cv2.IMREAD_GRAYSCALE)

# Compute DFT using FFT function
dft = np.fft.fft2(img)
dft_shifted = np.fft.fftshift(dft)

# Compute magnitude and phase
magnitude = np.abs(dft_shifted)
phase = np.angle(dft_shifted)

# Log transformation for visualization
magnitude_log = np.log1p(magnitude)

# Inverse FFT
ifft_shifted = np.fft.ifftshift(dft_shifted)
img_reconstructed = np.fft.ifft2(ifft_shifted).real

# Plot results
plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')

plt.subplot(2, 2, 2)
plt.imshow(magnitude_log, cmap='gray')
plt.title('DFT Magnitude Spectrum')

plt.subplot(2, 2, 3)
plt.imshow(phase, cmap='gray')
plt.title('DFT Phase Spectrum')

plt.subplot(2, 2, 4)
plt.imshow(img_reconstructed, cmap='gray')
plt.title('Reconstructed Image')

plt.show()
```

```python
img = cv2.imread('p5.jpg', cv2.IMREAD_GRAYSCALE)

dft = np.fft.fft2(img)
dft_shifted = np.fft.fftshift(dft)

magnitude = np.abs(dft_shifted)
phase = np.angle(dft_shifted)

magnitude_log = np.log1p(magnitude)

dft_ishift = np.fft.ifftshift(dft_shifted)
img_reconstructed = np.fft.ifft2(dft_ishift).real

plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')

plt.subplot(2, 2, 2)
plt.imshow(magnitude_log, cmap='gray')
plt.title('DFT Magnitude Spectrum')

plt.subplot(2, 2, 3)
plt.imshow(phase, cmap='gray')
plt.title('DFT Phase Spectrum')

plt.subplot(2, 2, 4)
plt.imshow(img_reconstructed, cmap='gray')
plt.title('Reconstructed Image')

plt.show()
```

```python
img = cv2.imread('p5.jpg', cv2.IMREAD_GRAYSCALE)

patch_8x8 = img[:8, :8]
patch_16x16 = img[:16, :16]

dft_8x8 = np.fft.fft2(patch_8x8)
dft_16x16 = np.fft.fft2(patch_16x16)

real_8x8, imag_8x8 = dft_8x8.real, dft_8x8.imag
real_16x16, imag_16x16 = dft_16x16.real, dft_16x16.imag

print(imag_8x8)
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.imshow(real_8x8, cmap='gray')
plt.title('DFT Real (8x8)')

plt.subplot(2, 2, 2)
plt.imshow(imag_8x8, cmap='gray')
plt.title('DFT Imaginary (8x8)')

plt.subplot(2, 2, 3)
plt.imshow(real_16x16, cmap='gray')
plt.title('DFT Real (16x16)')

plt.subplot(2, 2, 4)
plt.imshow(imag_16x16, cmap='gray')
plt.title('DFT Imaginary (16x16)')

plt.show()
```

```python
img1 = cv2.imread('p1.jpg', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('p2.jpg', cv2.IMREAD_GRAYSCALE)


h, w = img1.shape
img2 = cv2.resize(img2, (w, h))


dft1 = np.fft.fft2(img1)
dft2 = np.fft.fft2(img2)

dft1_shifted = np.fft.fftshift(dft1)
dft2_shifted = np.fft.fftshift(dft2)


mag1, phase1 = np.abs(dft1_shifted), np.angle(dft1_shifted)
mag2, phase2 = np.abs(dft2_shifted), np.angle(dft2_shifted)


dft_swapped_1 = mag2 * np.exp(1j * phase1) 
dft_swapped_2 = mag1 * np.exp(1j * phase2)  


dft_swapped_1 = np.fft.ifftshift(dft_swapped_1)
dft_swapped_2 = np.fft.ifftshift(dft_swapped_2)

img_reconstructed_1 = np.fft.ifft2(dft_swapped_1).real
img_reconstructed_2 = np.fft.ifft2(dft_swapped_2).real


plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.imshow(img1, cmap='gray')
plt.title('Original Image 1')

plt.subplot(2, 3, 2)
plt.imshow(np.log1p(mag1), cmap='gray')
plt.title('Magnitude Spectrum 1')

plt.subplot(2, 3, 3)
plt.imshow(phase1, cmap='gray')
plt.title('Phase Spectrum 1')

plt.subplot(2, 3, 4)
plt.imshow(img2, cmap='gray')
plt.title('Original Image 2')

plt.subplot(2, 3, 5)
plt.imshow(np.log1p(mag2), cmap='gray')
plt.title('Magnitude Spectrum 2')

plt.subplot(2, 3, 6)
plt.imshow(phase2, cmap='gray')
plt.title('Phase Spectrum 2')

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img_reconstructed_1, cmap='gray')
plt.title('Reconstructed Image (Mag2 + Phase1)')

plt.subplot(1, 2, 2)
plt.imshow(img_reconstructed_2, cmap='gray')
plt.title('Reconstructed Image (Mag1 + Phase2)')

plt.show()
```
