```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("p1.jpg")
plt.imshow(img)
# plt.axis('off') 
plt.show()
```

```python
img = cv2.resize(img, (800,800), interpolation = cv2.INTER_AREA)
plt.imshow(img)
plt.show()
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()

```

```python
blur = cv2.GaussianBlur(img_rgb, (7,7), cv2.BORDER_DEFAULT)
plt.imshow(blur)
plt.show()
```

```python
eroded = cv2.erode(img_rgb, (7,7), iterations = 3)
plt.imshow(eroded)
plt.show()
```

```python
cropped = img_rgb[50:200, 200:700]
plt.imshow(cropped)
```
