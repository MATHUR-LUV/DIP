```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
size = 8
arr = np.zeros((size, size))
print(arr)
```

```python
img = cv2.imread('p5.jpg', cv2.IMREAD_GRAYSCALE)
print(img)
```

```python
plt.plot(img)
```

```python
plt.imshow(img, cmap='gray')
plt.axis('off')
```

```python
def plot_dct_basis(img, size):
    fig, axes = plt.subplots(size, size, figsize=(5,5))
    for i in range(size):
        for j in range(size):
            arr = np.zeros((size, size))
            arr[i,j] = 1

            dct_basis = cv2.dct(np.float32(arr))
            axes[i,j].imshow(dct_basis, cmap='gray')
            axes[i,j].axis('off')
    plt.show()

def dct_2d(img):
    img_flt = np.float32(img)
    return cv2.dct(img_flt)
```

```python
dct_plot = plot_dct_basis(img, 8)

dct_2d_img = dct_2d(img)

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title('orig. image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(np.log(np.abs(dct_2d_img) + 1), cmap='gray')
plt.title('dct image')
plt.axis('off')

plt.show()
           
```
