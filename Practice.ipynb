```python
def bit_plane_slicing(image):
    bit_planes = []
    for i in range(8):
        bit_plane = (image >> i ) & 1;
        bit_plane = bit_plane * 255;
        bit_planes.append(bit_plane)
    return bit_planes;


img_gray = cv2.imread("p5.jpg", cv2.IMREAD_GRAYSCALE)
bit_planes = bit_plane_slicing(img_gray)

plt.figure(figsize=(12,12))
for i in range(8):
    plt.subplot(3,3,i+1)
    plt.imshow(bit_planes[i], cmap='gray')
    plt.title(f'bit_plane {i}')

plt.show()
```
