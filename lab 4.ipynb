```python
def histogram_equalization(image):
 
    hist, bins = np.histogram(image.flatten(), 256, [0, 256])
    print(bins)
    print(hist)
    cdf = hist.cumsum()
    print(cdf)
    cdf_normalized = (cdf - cdf.min()) * 255 / (cdf.max() - cdf.min())

    equalized_image = np.interp(image.flatten(), bins[:-1], cdf_normalized)

    return equalized_image.reshape(image.shape).astype(np.uint8)


img3 = cv2.imread('p2.jpg', cv2.IMREAD_GRAYSCALE)


equalized_image = histogram_equalization(img3)


plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.subplot(1, 2, 2)
plt.title('Equalized Image')
plt.imshow(equalized_image, cmap='gray')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def compute_histogram(image):
    hist = np.zeros(256, dtype=int)
    for pixel in image.flatten():
        hist[pixel] += 1
    return hist

def compute_cdf(hist):
    cdf = hist.cumsum()
    return cdf

def histogram_specification(source_image, reference_image):
    
    hist_source = compute_histogram(source_image)
    hist_reference = compute_histogram(reference_image)

    
    cdf_source = compute_cdf(hist_source)
    cdf_reference = compute_cdf(hist_reference)

   
    cdf_source_normalized = (cdf_source - cdf_source.min()) * 255 / (cdf_source.max() - cdf_source.min())
    cdf_reference_normalized = (cdf_reference - cdf_reference.min()) * 255 / (cdf_reference.max() - cdf_reference.min())

   
    lut = np.zeros(256, dtype=np.uint8)
    for i in range(256):
       
        idx = np.abs(cdf_reference_normalized - cdf_source_normalized[i]).argmin()
        lut[i] = idx

  
    specified_image = lut[source_image]

    return specified_image

source_image = cv2.imread('p2.jpg', cv2.IMREAD_GRAYSCALE)
reference_image = cv2.imread('p1.jpg', cv2.IMREAD_GRAYSCALE)

specified_image = histogram_specification(source_image, reference_image)


plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.title('Source Image')
plt.imshow(source_image, cmap='gray')
plt.subplot(1, 3, 2)
plt.title('Reference Image')
plt.imshow(reference_image, cmap='gray')
plt.subplot(1, 3, 3)
plt.title('Specified Image')
plt.imshow(specified_image, cmap='gray')
plt.show()
 
```

```python
# Contrast Limited Adaptive Histogram Equalization (CLAHE) 
```

```python

def clip_histogram(hist, clip_limit):
 
    excess = np.maximum(0, hist - clip_limit)
    clipped_hist = hist - excess
   
    excess_sum = np.sum(excess)
    clipped_hist += excess_sum // 256
    return clipped_hist


def apply_clahe_to_region(region, clip_limit):
  
    hist = compute_histogram(region)

    clipped_hist = clip_histogram(hist, clip_limit)


    cdf = compute_cdf(clipped_hist)

    cdf_normalized = ((cdf - cdf.min()) * 255) / (cdf.max() - cdf.min())

    equalized_region = np.interp(region.flatten(), np.arange(256), cdf_normalized)
    return equalized_region.reshape(region.shape).astype(np.uint8)


def clahe(image, tile_size=8, clip_limit=40):

    h, w = image.shape

    result_image = np.zeros_like(image)

    for i in range(0, h, tile_size):
        for j in range(0, w, tile_size):
          
            x_end = min(i + tile_size, h)
            y_end = min(j + tile_size, w)
            tile = image[i:x_end, j:y_end]

            equalized_tile = apply_clahe_to_region(tile, clip_limit)

            result_image[i:x_end, j:y_end] = equalized_tile

    return result_image


cl = cv2.imread('p2.jpg', cv2.IMREAD_GRAYSCALE)


clahe_image = clahe(cl, tile_size=16, clip_limit=100)


plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cl, cmap='gray')
plt.subplot(1, 2, 2)
plt.title('CLAHE Image')
plt.imshow(clahe_image, cmap='gray')
plt.show()


```
